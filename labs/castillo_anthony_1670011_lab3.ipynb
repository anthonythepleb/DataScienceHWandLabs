```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
rcParams['figure.figsize'] = 15, 5
sns.set_style('darkgrid')
```

```python
x = np.linspace(0,20)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("My graph title")
plt.xlabel("my x-axis label")
plt.ylabel("my y-axis label")
plt.show()
```

```python
plt.plot(x, np.sin(x), label = "y=sin(x)")
plt.plot(x, np.cos(x), label = "y=cos(x)")
plt.title("Sine and Cosine Functions")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,np.sin(x))
axes[0].set_title("y=sin(x)")

axes[1].plot(x,np.cos(x))
axes[1].set_title("y=cos(x)")
plt.show()
```

```python
x = np.linspace(1,21)

fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (14,8))
axes[0][0].plot(x,x**2)
axes[0][1].plot(x,x)
axes[1][0].plot(x,np.log(x))
axes[1][1].plot(x,x**5)

plt.show()
```

```python
x = np.linspace(0,5,11)

fig, axes = plt.subplots(nrows = 1, ncols = 3)

axes[0].plot(x, x**2, x, x**4)
axes[0].set_title("default axis range")

axes[1].plot(x,x**2,x,x**4)
axes[1].axis("tight")
axes[1].set_title("tight axis range")

axes[2].plot(x,x**2,x,x**4)
axes[2].set_ylim([0,40])
axes[2].set_xlim([2, 3])
axes[2].set_title("custom axes range")
plt.show()
```

```python
x = pokemon_df["Name"][:4]
y1 = pokemon_df["HP"][:4]
y2 = pokemon_df["Attack"][:4]
y3 = pokemon_df["Defense"][:4]
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1, sharey = True)

axes[0].plot(x,y1, "ko-")
axes[0].set(title = "3 subplots", ylabel = "HP")

axes[1].plot(x, y2, "r.-")
axes[1].set(title = "Pokemon", ylabel = "Attack")

axes[2].plot(x,y3, ":")
axes[2].set(xlabel = "Pokemon", ylabel = "Defense")
plt.show()
```
